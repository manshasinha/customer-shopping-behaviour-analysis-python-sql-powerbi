## Treating Null Values

#### SIZE


```python
#size
df['Size'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.groupby('Item Purchased')['Size'].unique()
```

```python
df.groupby('Category')['Item Purchased'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#TREATING NULL VALUES/ REPLACING ALL VALUES IN ELECTRONICS CATEGORY
df.loc[df['Category']=='Electronics','Size']="not applicable"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#TREATING NULL VALUES/ REPLACING ALL VALUES IN Accessories CATEGORY
df.loc[df['Category']=='Accessories','Size']="free size"
```

```python
#TREATING NULL VALUES/ REPLACING ALL VALUES IN Footwear CATEGORY
df.loc[df['Category']=='Footwear','Size']="not availabe"
```

```python
# Treating null values or replacing all values in the Clothing category
df.loc[(df['Category'] == 'Clothing') & (df['Size'].isnull()), 'Size'] = df[df['Category'] == 'Clothing']['Size'].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.isnull().sum()
```

### REVIEW RATING

```python
df.head()
```

```python
#finding mean of review rating
df['Review Rating'].mean()
```

```python
#finding median of review rating
df['Review Rating'].median()
```

```python
#finding mode of review rating
df['Review Rating'].mode()
```

```python
df.groupby('Item Purchased')['Review Rating'].mean()
```

```python
# Filling null values of Review Ratings with the mean of each Item Purchased
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
#rounding off review ratings to 2 decimal places
df['Review Rating']=df['Review Rating'].round(2)
```

```python
df.head()
```

```python
df.isnull().sum()
```

### PREVIOUS PURCHASES


```python
df['Previous Purchases']=df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

### Purchase Amount (USD)

```python
#replacing missing purchase prices of items with their average price
df.groupby('Item Purchased')['Purchase Amount (USD)'].mean()
```

```python
# Filling null values of Purchase Amount (USD) with the mean of each Item Purchased
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```
